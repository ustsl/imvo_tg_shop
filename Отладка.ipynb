### Telegram imvo shop

1. Блок - администратор. Админ при публикации товара пишет сообщение. если в сообщение есть поле Цена: 21312312 руб, Номер: 342342344, то записываем в базу данных поля "номер и цена"

<br>

2. База данных - обычный JSON, где пишем номер заказа и его цену
    
<br>

3. Если человек пишет номер заказа в ленте, то мы ему отдаем сообщение (при успехе) - ваш заказ передан, при неуспехе - вы ввели несуществующий, или уже неактуальный номер заказа
    
<br>

4. Передаем номер заказа всем администраторам магазина
